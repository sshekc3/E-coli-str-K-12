```python
from Bio import SeqIO

# Read reference sequence
ref = SeqIO.read("sequence.fasta", "fasta") # Changed from reference.fasta
print("Reference ID:", ref.id)
print("Length of reference sequence:", len(ref.seq))
print("First 50 bases of reference:", ref.seq[:50])

# Read sample sequence
sample = SeqIO.read("sequence.fasta", "fasta") # Changed from sample.fasta
print("\nSample ID:", sample.id)
print("Length of sample sequence:", len(sample.seq))
print("First 50 bases of sample:", sample.seq[:50])
```

```python
import pandas as pd
```

```python
# Count each base
base_counts = {
    "A": ref.seq.count("A"),
    "T": ref.seq.count("T"),
    "G": ref.seq.count("G"),
    "C": ref.seq.count("C")
}

df_counts = pd.DataFrame(list(base_counts.items()), columns=["Base", "Count"])
print(df_counts)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.bar(df_counts["Base"], df_counts["Count"], color=['green','red','blue','orange'])
plt.title("Nucleotide composition")
plt.ylabel("Count")
plt.show()
```
